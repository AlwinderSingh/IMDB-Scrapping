```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_data(pageno):
    url=('https://www.imdb.com/search/title/?release_date=2020-01-01,2020-12-31&sort=num_votes,desc&start='+str(pageno)+'&ref_=adv_nxt')
    r=requests.get(url)
    print(url)
    html_content=r.content
    soup=BeautifulSoup(html_content)
    all=[]
    #print(html_content)
    for d in soup.find_all('div',attrs={'class':'lister-item-content'}):
        name=d.find('a')
        imdb=d.find('div',attrs={'class':'inline-block ratings-imdb-rating'})
        imdb=imdb.find('strong')
        description=d.find('p',attrs={'class':'text-muted'})
        description=description.find_next_sibling('p',attrs={'class':'text-muted'})
        votes=d.find('p',attrs={'class':'sort-num_votes-visible'})
        votes=votes.find('span',attrs={'name':'nv'})
        info=d.find_all('p')[-2]
        director=info.find('a')
        stars=info.find_next_siblings('a')
        metascore=d.find('div',attrs={'class':'inline-block ratings-metascore'})
        #metascore=metascore.find('span',attrs={'class':'metascore  favorable'})
        mgenre=d.find('p')#,attrs={'class':'text-muted '})
        tduration=mgenre.find('span',attrs={'class':'runtime'})
        mgenre=mgenre.find('span',attrs={'class':'genre'})
        #tduration=d.find('p',attrs={'class':'text-muted '})
        all1=[]
        #print(info.text)
        if name is not None:
            all1.append(name.text)
        else:
            all1.append('unknown')
        if imdb is not None:
            all1.append(imdb.text)
        else:
            all1.append('unknown')
        if description is not None:
            all1.append(description.text[1:])
        else:
            all1.append('unknown')
        if votes is not None:
            all1.append(votes.text)
        else:
            all1.append('unknown')
        if info is not None:
            x=info.text.replace('\n','').replace('|','').replace('Director:','').replace('Directors:','').split('Stars:')
            x[0]=x[0].lstrip()
            if(x[0]==''):
                x[0]='unknown'
            all1.append(x[0])
            if(len(x)>1):
                all1.append(x[1])
            else:
                all1.append('unknown')
        else:
            all1.append('unknown')
        if metascore is not None:
            all1.append(metascore.text.replace('\n','').replace('Metascore','').lstrip())
        else:
            all1.append('unknown')
        if mgenre is not None:
            all1.append(mgenre.text.replace('\n','').lstrip())
            #all1.append(mgenre.text.split('|')[-1].replace('\n','').lstrip())
        else:
            all1.append('unknown')
        if tduration is not None:
         #   all1.append(mgenre.text.split('|')[-2].replace('\n','').lstrip())
            all1.append(tduration.text)
        else:
            all1.append('unknown')
        #print(all1)
        all.append(all1)
        
    return all
```

```python
import pandas as pd
results = []
no_pages=5
for i in range(1, 1001, 50):
    results.append(get_data(i))
flatten = lambda l: [item for sublist in l for item in sublist]
df = pd.DataFrame(flatten(results),columns=['Movie_Name','IMDB_Rating','Descriptions','Votes','Director(s)','Stars','Metascore','Movie_Genre','Total_Duration'])
```
